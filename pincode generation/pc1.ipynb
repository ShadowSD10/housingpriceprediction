```python
import pandas as pd
from geopy.geocoders import Nominatim
import time
```

```python
df = pd.read_csv("latlong.csv")
```

```python
df.head()
```

```python
col = ['latitude', 'longitude']
df = df[col]
```

```python
df.head()
```

```python
df = df.assign(pincode='')
```

```python
df
```

```python
# Initialize a Nominatim geocoder object
geolocator = Nominatim(user_agent="my_geocoder")

# Initialize an empty list to store pin codes
pincode_list = []

# Loop through the latitude and longitude data from the df_encod dataframe
for index, row in df.iloc[0:1000].iterrows():
    latitude = row['latitude']
    longitude = row['longitude']

    try:
        # Get location information (address) for the given latitude and longitude
        location = geolocator.reverse((latitude, longitude), exactly_one=True)
        if location:
            # Extract pin code from location address
            pin_code = location.raw['address'].get('postcode')
            pincode_list.append(pin_code)
        else:
            pincode_list.append(None)
    except Exception as e:
        print(f"Error: {e}")
        pincode_list.append(None)

# Create a new dataframe with 'pincode' column
pincode_df = pd.DataFrame({'pincode': pincode_list})

# Print the new dataframe
print(pincode_df)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import time

# Initialize Nominatim geocoder
geolocator = Nominatim(user_agent="geoapiExercises")

# Function to calculate pincode from latitude and longitude
def calculate_pincode(latitude, longitude):
    try:
        # Get location information (address) for the given latitude and longitude
        location = geolocator.reverse((latitude, longitude), exactly_one=True)
        if location:
            # Extract pin code from location address
            pin_code = location.raw['address'].get('postcode')
            return pin_code
        else:
            return None
    except Exception as e:
        print(f"Error: {e}")
        return None

# Loop through the latitude and longitude data from the df_encod dataframe
for index, row in df.iterrows():
    if pd.isna(row['pincode']) or row['pincode'] == '':
        latitude = row['latitude']
        longitude = row['longitude']

        try:
            # Calculate pin code from latitude and longitude
            pincode = calculate_pincode(latitude, longitude)
            if pincode:
                df.at[index, 'pincode'] = pincode
            time.sleep(1)  # Throttle requests to avoid exceeding usage limits
        except Exception as e:
            print(f"Error: {e}")

print(df)
```
